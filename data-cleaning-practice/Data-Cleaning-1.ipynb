### Cleaning phone number column

all rows of phone number are different type

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True) # remove everthing except this
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x)) # change all rows of phone no to string
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--','')
df
```

### Cleaning address column


```python
for x in df.index:
    print(type(df['Address'][x]))
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',',n=2, expand=True)
df
```

### Clean paying customer column

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
# df['Paying Customer'].str.replace('[^\b(Yes|No|N|N/a)\b.*]', 'Yes', regex=True) # remove everthing except this
# df['Paying Customer']
# df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex=True) # remove everthing except this
```

### Clean Do not Contact column

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No','N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df
```

```python
df = df.replace('N/a','')
# df.replace('NaN','')
df = df.fillna('')
df
```
